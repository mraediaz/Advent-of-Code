```python
import re
import numpy as np
# with open('./example_input.txt', 'r') as f:
with open('./input.txt', 'r') as f:
    lines = [line.strip() for line in f.readlines() if line.strip()]

presents = []
trees = []
lookup = []
i = 0
n = len(lines)

while i < n:
    line = lines[i]
    # presents
    if re.fullmatch(r'\d+:', line):
        pattern = []
        j = i
        i += 1
        while i <= j+3:
            row = lines[i]
            pattern.append([(1 if c == '#' else 0) for c in row])
            i += 1
        presents.append(pattern)

    # Match grid definition: "4x4:"
    elif re.match(r'\d+x\d+:', line):
        match = re.match(r'(\d+)x(\d+): ((\d+\s?)+)', line)
        rows, cols = int(match.group(1)), int(match.group(2))
        
        tree = np.zeros((rows,cols))
        trees.append(tree)

        shapes_needed = match.group(3).split(' ')
        lookup.append(shapes_needed)
        i += 1
```

```python
regions = len(trees)
for i, t in enumerate(trees):
    (m,n) = t.shape
    tree_area = m*n
    
    area_presents = 0
    for idx, count in enumerate(lookup[i]):
        if int(count) > 0:
            # print("we need", int(count), "presents of type ", idx)
            (m_, n_) = np.shape(presents[idx])
            area_presents += (int(count) * (m_*n_))
    if area_presents > tree_area:
        regions -=1
            
print("part 1:", regions)        
```
