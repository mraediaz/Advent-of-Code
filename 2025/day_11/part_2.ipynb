```python
paths = {"out": []}
# with open("input.txt", "r") as f:
with open("example_input_2.txt", "r") as f:
    for line in f:
        devices = line.strip().split(' ')
        paths[devices[0][:-1]] = devices[1:]
```

```python
# %%timeit 30.6 μs ± 654 ns per loop (mean ± std. dev. of 7 runs, 10,000 loops each)
# part 2 
from functools import lru_cache

# def dfs_count(start, finish, tree):
#     if start == finish:
#         return 1
#     total = 0
#     for child_node in tree[start]:
#         total += dfs_count(child_node, finish, tree)
#     return total

def dfs_count(start, finish, tree):
    @lru_cache(maxsize=None)
    def dfs(node):
        if node == finish:
            return 1
        total = 0
        for child in tree.get(node, []):
            total += dfs(child)
        return total
    return dfs(start)

# svr->dac->fft->out
a = dfs_count('svr', 'dac', paths)
b = dfs_count('dac', 'fft', paths)
c = dfs_count('fft', 'out', paths)

# svr->fft->dac->out
d = dfs_count('svr', 'fft', paths)
e = dfs_count('fft', 'dac', paths)
f = dfs_count('dac', 'out', paths)

print("part 2: ", (a*b*c) + (d*e*f))
```
