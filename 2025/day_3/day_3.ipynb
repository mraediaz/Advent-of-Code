```python
with open("input.txt", "r") as f:
    instructions = [line.strip() for line in f if line.strip()]
```

```python
#part 1
sum = 0
for bank in instructions:
    bank_sum = first_index = 0
    first = bank[:-1]

    for battery in first:
        first_max = max(first)
        first_index = bank.index(first_max)

    second = bank[first_index+1:]
    for battery in second:
        second_max = max(second)
        bank_sum = (int(first_max)*10)+int(second_max)
    # print(f"{bank}: {bank_sum}")
    sum+=bank_sum
print(sum)
    
```

```python
#part 2
sum = 0
for bank in instructions:
    bank_sum = array_start = 0
    sub_array = []
    
    for i in range(12):
        batteries_needed = 12 - i - 1
        sub_array_index = len(bank) - batteries_needed

        max_battery = max(bank[array_start:sub_array_index])
        max_index = bank.index(max_battery, array_start)

        sub_array.append(max_battery)
        array_start = max_index + 1

        bank_sum =  int(''.join(sub_array))   
    sum+=bank_sum
print(sum)
```
