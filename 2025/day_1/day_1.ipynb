```python
import polars as pl
import re
```

```python
df = pl.read_csv("input.txt", has_header=False, new_columns=["instruction"]) 
# df = pl.read_csv("example_input.txt", has_header=False, new_columns=["instruction"]) 
```

```python
%%timeit
#day 1
cnt_0 = 0
place = 50

for instruction in df["instruction"]:
    instruct = re.split(r'(\w)(\d+)', instruction)
    # print('===========')
    # print(instruct)
    operator = instruct[1]
    num = int(instruct[2])

    if operator == 'R':
        place += num
    elif operator == 'L':
        place -= num

    place = place % 100    
    if place == 0:
        cnt_0 += 1 
print(cnt_0)
```

```python
# %%timeit
cnt_0 = 0
place = 50

for instruction in df["instruction"]:
    instruct = re.split(r'(\w)(\d+)', instruction)
    # print('===========')
    # print(instruct)
    operator = instruct[1]
    num = int(instruct[2])

    starting_at_zero = place != 0

    if operator == 'R':
        place += num
    elif operator == 'L':
        place -= num


    if place == 0:
        cnt_0 += 1
    elif place >= 100:
        div = place // 100
        cnt_0 += div
    elif place < 0:
        # print(f"pointing_at {place}, starting_at_zero {starting_at_zero}: {place // -100 + starting_at_zero}")
        div = place // -100 + starting_at_zero
        cnt_0 += div

    place = place % 100
     
print(cnt_0)
```
