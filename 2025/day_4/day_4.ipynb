```python
# part 2
def day_4(grid):
    new_grid = [list(row) for row in grid]
    rows = len(grid)
    cols = len(grid[0]) - 1

    # right, left, down, up, and all diagonals
    directions = [
        (0, 1, "right"),   # →
        (0, -1, "left"),   # ←
        (1, 0, "down"),    # ↓
        (-1, 0, "up"),     # ↑
        (1, 1, "diagonal right-down"),    # ↘
        (-1, -1, "diagonal left-up"),     # ↖
        (-1, 1, "diagonal right-up"),     # ↗
        (1, -1, "diagonal left-down")     # ↙
    ]

    cnt_rolls = 0
    for row in range(rows):
        for col in range(cols):
            if grid[row][col] == '.':
                continue
            # print(f"======{row}:{col}")
            cnt_neighbors= 0
            for delta_row, delta_col, direction in directions:
                neighbor_row = row+delta_row
                neighbor_col = col+delta_col
                if not (0 <= neighbor_row < rows and 0 <= neighbor_col < cols):
                    continue
                # print(neighbor_row, neighbor_col)
                if grid[neighbor_row][neighbor_col] == '@':
                    cnt_neighbors+=1
            # print(f"={cnt_neighbors}")

            if cnt_neighbors < 4:
                new_grid[row][col] = '.'
                cnt_rolls+=1
    return (new_grid,cnt_rolls)



new_grid,cnt_rolls = day_4(grid)
full_counter = cnt_rolls
print(f"part 1: {full_counter}")

while cnt_rolls > 0:
    new_grid, cnt_rolls = day_4(new_grid)
    full_counter+=cnt_rolls
print(f"part 1: {full_counter}")
    
```
