```python
import polars as pl
import re
```

```python
input_df = pl.read_csv("input.txt", has_header=False, new_columns=["instruction"]) 
```

```python
df = (
    input_df.unpivot()
    .select([
        pl.col("value"),
        pl.col("value").str.split("-").list.get(0).cast(pl.Int64).alias("first_id"),
        pl.col("value").str.split("-").list.get(1).cast(pl.Int64).alias("last_id")
    ])
)
```

```python
%%timeit
#part 1
invalid_id_cnt = 0
for row in df.iter_rows():
    for id in range(row[1],row[2]+1):
        id_str = str(id)
        mid = len(id_str) // 2
        invalid_id_cnt+=id if id_str[:mid] == id_str[mid:] else 0
print(invalid_id_cnt)
```

```python
%%timeit
#part 2
invalid_id_cnt = 0
for row in df.iter_rows():
    for id in range(row[1],row[2]+1):
        id_str = str(id)

        match_parts = re.match(r"(\d+?)\1+$", id_str)
        if match_parts:
            invalid_id_cnt+=id
            # repeating_part = match_parts.group(1)
            # print(f"id: {id}, repeating_part: {repeating_part}, invalid_id_cnt: {invalid_id_cnt}")

        ''' 
        # second regex expression
        match = re.search(r'^(\d+?)\1+$', id_str)
        if match:
            print(f"id: {id_str}, {match.group(1)}")
            invalid_id_cnt+=id
        '''
print(invalid_id_cnt)
```
