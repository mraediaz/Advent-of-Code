```python
import numpy as np
```

```python
jb = []
with open('input.txt', 'r') as file:
    for line in file:
        line = line.strip()
        if line: 
            arr = np.array(list(map(int, line.split(','))))
            jb.append(arr)
```

```python
ed = []
for i in range(len(jb)):
    X = jb[i]
    for j in range(i+1,len(jb)):
        if i != j:
            ed.append([jb[i], jb[j], np.linalg.norm(jb[i] - jb[j])])

sorted_ed = sorted(ed, key=lambda x: x[2])
```

```python
def day_8(cnt):
    # Mapping from junction to the circuit (set) it belongs to
    junction_to_circuit = {}
    all_circuits = []  # List of all circuits (each is a set of junctions)

    for i in range(cnt):
        # Get the two junctions and distance (already sorted by distance)
        junc1, junc2, _ = sorted_ed[i]
        junc1 = tuple(junc1)
        junc2 = tuple(junc2)

        in_circuit_1 = junc1 in junction_to_circuit
        in_circuit_2 = junc2 in junction_to_circuit

        if not in_circuit_1 and not in_circuit_2:
            # Both junctions are new: create a new circuit
            new_circuit = {junc1, junc2}
            all_circuits.append(new_circuit)
            junction_to_circuit[junc1] = new_circuit
            junction_to_circuit[junc2] = new_circuit

        elif in_circuit_1 != in_circuit_2:
            # One is in a circuit, the other isn't
            existing_junc = junc1 if in_circuit_1 else junc2
            new_junc = junc2 if in_circuit_1 else junc1
            circuit = junction_to_circuit[existing_junc]
            circuit.add(new_junc)
            junction_to_circuit[new_junc] = circuit

        elif in_circuit_1 and in_circuit_2:
            # Both are already in circuits
            circuit_1 = junction_to_circuit[junc1]
            circuit_2 = junction_to_circuit[junc2]

            if circuit_1 != circuit_2:
                # Merge the two circuits
                merged_circuit = circuit_1 | circuit_2

                # Update all junctions to point to the merged circuit
                for junction in merged_circuit:
                    junction_to_circuit[junction] = merged_circuit

                # Update the list of circuits
                all_circuits.remove(circuit_1)
                all_circuits.remove(circuit_2)
                all_circuits.append(merged_circuit)

        # Part 2
        if cnt == len(sorted_ed):
            if len(all_circuits) == 1 and len(all_circuits[0]) == len(jb):
                return junc1[0] * junc2[0]  # Product of x-coordinates

    # Part 1
    if cnt == 1000:
        all_circuits.sort(key=len, reverse=True)
        top_three = all_circuits[:3]
        return len(top_three[0]) * len(top_three[1]) * len(top_three[2])
```

```python
print("part 1:", day_8(1000))
print("part 2:", day_8(len(sorted_ed)))
    
```
